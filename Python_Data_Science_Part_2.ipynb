<h4> Matplotlib </h4>

```
import matplotlib.pyplot as plt
#This can avoid show
%matplotlib inline
```

```
import numpy as np
```

```
x=np.arange(10,20)
y=np.arange(40,50)
```

```
x
```

```
plt.scatter(x,y,c='g')
plt.xlabel("X values")
plt.ylabel("Y value")
plt.title("First Scatter Plot")
```

```
plt.plot(x,y,'ro--')
```

```
#2 rows and 2 columns
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```
#Sin wave
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```
x=np.arange(0,2*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
```
